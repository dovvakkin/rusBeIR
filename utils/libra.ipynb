```python
with open('dataset.jsonl', 'w', encoding='utf-8') as f:
    for idx, record in enumerate(df.to_dict(orient='records')):
        record['_id'] = idx  
        f.write(json.dumps(record, ensure_ascii=False) + '\n')
```

```python
corpus_dict = {}
queries = []
tsv_pairs = ['query_id\tcorpus_id\tscore']

text_id_map = {}
next_corpus_id = 0
next_query_id = 100

for _, item in df.iterrows():
    text = item["context"]
    question = item["input"]
    
    if text not in text_id_map:
        text_id_map[text] = next_corpus_id
        corpus_dict[next_corpus_id] = {"_id": next_corpus_id, 'title': '', "text": text}
        next_corpus_id += 1
    
    corpus_id = text_id_map[text]
    
    query_id = next_query_id
    queries.append({"_id": query_id, "text": question})
    
    tsv_pairs.append(f"{query_id}\t{corpus_id}\t1")
    
    next_query_id += 1

with open('corpus.jsonl', 'w', encoding='utf-8') as f:
    for entry in corpus_dict.values():
        f.write(json.dumps(entry, ensure_ascii=False) + '\n')

with open('queries.jsonl', 'w', encoding='utf-8') as f:
    for entry in queries:
        f.write(json.dumps(entry, ensure_ascii=False) + '\n')

with open('test.tsv', 'w', encoding='utf-8') as f:
    for pair in tsv_pairs:
        f.write(pair + '\n')

print("Файлы успешно созданы.")
```

```python
dataset = load_dataset('json', data_files='dataset.jsonl')
```

```python
def split_into_paragraphs(example):
    paragraphs = [p.strip() for p in example['context'].split('\n\n') if len(p.strip()) > 20]
    new_examples = []
    
    for idx, paragraph in enumerate(paragraphs):
        if paragraph.strip():  
            new_example = example.copy()  
            new_example['context'] = paragraph.strip()  
            new_example['_id'] = f"{example['_id']}/{idx}"
            new_examples.append(new_example)
    
    return new_examples
```

```python
expanded_dataset = []
for doc in dataset['train']:
    expanded_dataset.append(split_into_paragraphs(doc))
    

expanded_ds = np.concatenate(expanded_dataset).tolist()
```

```python
with open('dataset_abstracted_filtered.jsonl', 'w') as f:
    for el in expanded_ds:
        f.write(json.dumps(el, ensure_ascii=False) + '\n')
```

```python
df = pd.DataFrame(expanded_ds)
```

```python
df
```

```python
import json

corpus_dict = {}
queries_dict = {} 
tsv_pairs = ['query_id\tcorpus_id\tscore']

next_query_id = 100

for _, item in df.iterrows():
    text_id = item["_id"]  
    text = item["context"]
    question = item["input"]
    
    if text_id not in corpus_dict:
        corpus_dict[text_id] = {"_id": text_id, 'title': '', "text": text}
    
    if question not in queries_dict:
        queries_dict[question] = {"_id": next_query_id, "text": question}
        query_id = next_query_id
        next_query_id += 1
    else:
        query_id = queries_dict[question]["_id"]  
    
    tsv_pairs.append(f"{query_id}\t{text_id}\t1")

with open('corpus.jsonl', 'w', encoding='utf-8') as f:
    for entry in corpus_dict.values():
        f.write(json.dumps(entry, ensure_ascii=False) + '\n')

with open('queries.jsonl', 'w', encoding='utf-8') as f:
    for entry in queries_dict.values():
        f.write(json.dumps(entry, ensure_ascii=False) + '\n')

with open('test.tsv', 'w', encoding='utf-8') as f:
    for pair in tsv_pairs:
        f.write(pair + '\n')

print("Файлы успешно созданы.")
```
